**Handling Missing Values**

```python
import pandas as pd
import numpy as np
from pathlib import Path

# Load your raw data
# try common locations, then search project if not found
candidates = [
    Path('data/raw/creditcard.csv'),
    Path('../data/raw/creditcard.csv'),
    Path('data/creditcard.csv'),
    Path('creditcard.csv')
]
csv_path = next((p for p in candidates if p.exists()), None)
if csv_path is None:
    found = list(Path('.').rglob('creditcard.csv'))
    csv_path = found[0] if found else None

if csv_path is None:
    raise FileNotFoundError(
        "creditcard.csv not found. Place it under data/raw/ or upload it to the notebook working directory."
    )

df = pd.read_csv(csv_path)

missing_pct = df.isnull().mean() * 100
print("Missing Data Percentage:\n", missing_pct[missing_pct > 0])

if 'transaction_amount' in df.columns:
    median_amount = df['transaction_amount'].median()
    df['transaction_amount'] = df['transaction_amount'].fillna(median_amount)

# 2. Impute a categorical feature with a placeholder
if 'device_type' in df.columns:
    df['device_type'] = df['device_type'].fillna('Unknown')
```
